### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
#unir tablas en una sola y generar id unicos para el total de registros concatenados
df_total = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
dfVentasxTienda = df_total.groupby('Tienda')['Precio'].sum().reset_index()
dfVentasxTienda
```

```python
dfVentasxTienda.columns = ['Tienda', 'Ventas Totales']
dfVentasxTienda = dfVentasxTienda.sort_values(by='Ventas Totales', ascending=False)

import matplotlib.pyplot as plt

# generar el gráfico de barras para estos datos
plt.figure(figsize=(10,6))
plt.bar(dfVentasxTienda['Tienda'], dfVentasxTienda['Ventas Totales'], color='blue')
plt.title("Ventas Totales por tienda")
plt.show()


dfVentasxTienda.style.format({'Ventas Totales': '${:,.0f}'})
```

```python
df_calificacion=df_total.groupby('Tienda')['Calificación'].mean().reset_index()
df_calificacion=df_calificacion.rename(columns={'Calificación':'Calificación Promedio'})
df_calificacion
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_calificacion['Tienda'], df_calificacion['Calificación Promedio'],
         marker='o', linestyle='-', color='gold', markersize=12, linewidth=2)
plt.title('Índice de Satisfacción del Cliente por Tienda', fontsize=14)
plt.ylabel('Calificación (1 a 5 Estrellas)', fontsize=12)
plt.xlabel('Sucursales', fontsize=10)
#plt.ylim(0, 5.5) # Fijamos el límite para que se vea la escala real de 1 a 5
plt.grid(axis='y', linestyle='--', alpha=0.5)
```

```python
# 1. Agrupamos por Tienda y Categoría para contar las ventas
conteo_ventas = df_total.groupby(['Tienda', 'Categoría del Producto']).size().reset_index(name='Cantidad')

# 2. Ordenamos por cantidad de mayor a menor
conteo_ventas = conteo_ventas.sort_values(['Tienda', 'Cantidad'], ascending=[True, False])

# 3. nos quedamos solo con la primera fila de cada tienda
top_categoria_por_tienda = conteo_ventas.groupby('Tienda').head(1)

top_categoria_por_tienda
```

```python
# 1. Agrupamos por Tienda y Categoría para sumar las ventas
suma_ventas = df_total.groupby(['Tienda', 'Categoría del Producto'])['Precio'].sum().reset_index()

# Aquí es donde le pones el nombre profesional
suma_ventas = suma_ventas.rename(columns={'Precio': 'Total'})

# 2. Ordenamos por suma de mayor a menor
suma_ventas = suma_ventas.sort_values(['Tienda', 'Total'], ascending=[True, False])

# 3. nos quedamos solo con la primera fila de cada tienda
top_venta_por_tienda = suma_ventas.groupby('Tienda').head(1)

# Extra: Aplica el formato de moneda para que se vea genial
top_venta_por_tienda.style.format({'Total': '${:,.0f}'})

```

# 3. Calificación promedio de la tienda


```python
# 1. Agrupamos por Tienda y calculamos el promedio de Calificación
df_calificacion = df_total.groupby('Tienda')['Calificación'].mean().reset_index()

# 2. Renombramos la columna para que sea clara
df_calificacion = df_calificacion.rename(columns={'Calificación': 'Calificación Promedio'})

# 3. Ordenamos de mejor a peor calificación (Descendente)
df_calificacion = df_calificacion.sort_values(by='Calificación Promedio', ascending=False)

# 4. Mostramos con formato de un solo decimal (para que no salgan muchos números)
df_calificacion.style.format({'Calificación Promedio': '{:.1f} ⭐'})
```


#Informe Estratégico de Optimización de Negocio: Caso Don Juan

## 1. Introducción

El presente análisis tiene como propósito evaluar el desempeño operativo y comercial de las cuatro sucursales del Sr. Juan. El objetivo final es identificar cuál de las tiendas presenta el menor rendimiento integral para proceder con su venta, liberando así capital para la inversión en un nuevo modelo de negocio. Para este diagnóstico, se han cruzado variables críticas como ingresos totales, satisfacción del cliente, eficiencia logística (costos de envío) y rotación de inventario por categorías.

---

## 2. Desarrollo del Análisis

### A. Rendimiento Financiero y Volumen de Ventas

A través del gráfico de **Ventas Totales por Tienda**, observamos una tendencia clara. Mientras que la **Tienda_1** lidera el mercado superando los $1,150 millones, la **Tienda_4** se posiciona consistentemente en el último lugar. Esta brecha de ingresos no es aislada; refleja una menor tracción de clientes en dicha ubicación.

### B. Análisis de Categorías e Inventario

El análisis de **Distribución de Ventas por Categoría** revela que el "corazón" del negocio son los **Electrónicos**.

* **Fortaleza:** Todas las tiendas dependen de la tecnología para su flujo de caja.
* **Debilidad:** La **Tienda_4** muestra la mayor acumulación de "productos menos vendidos" en categorías secundarias, lo que sugiere que su ubicación o perfil de cliente no es apto para diversificar el inventario, generando costos de almacenamiento innecesarios.

### C. Calidad de Servicio (Satisfacción del Cliente)

El **Índice de Satisfacción** (Calificación Promedio) es el indicador de salud a largo plazo.

* Las tiendas 1 y 2 mantienen promedios competitivos.
* La **Tienda_4** presenta las calificaciones más erráticas o bajas, lo que indica problemas operativos, mala atención o deficiencias en el servicio post-venta. Una baja calificación aumenta el costo de adquisición de clientes y daña la reputación de la marca global de Don Juan.

### D. Eficiencia Logística (Coste de Envío)

Al analizar el **Coste de Envío Promedio**, se identificó que la **Tienda_4** no logra economías de escala. Sus costos logísticos por pedido tienden a ser proporcionalmente más altos en relación con su volumen de ventas, reduciendo el margen de utilidad neta en comparación con la Tienda_1, donde la logística está optimizada por el alto volumen.

---

## 3. Conclusión y Recomendación Final

**Recomendación:** Se aconseja formalmente al Sr. Juan proceder con la **venta de la Tienda_4**.

### Justificación de la decisión:

1. **Menor Aporte Económico:** Es la sucursal con ingresos totales más bajos, representando el eslabón más débil de la cadena financiera.

2. **Riesgo de Reputación:** Sus bajas calificaciones promedio sugieren que los recursos invertidos en "salvar" la tienda podrían ser mejor aprovechados en un nuevo emprendimiento.

3. **Optimización de Capital:** La venta de la Tienda_4 permitirá a Don Juan reinvertir en la compra de mayor inventario de **Electrónicos** (la categoría estrella) para las Tiendas 1 y 2, maximizando así el retorno sobre la inversión (ROI).
